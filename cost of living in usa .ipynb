# Introduction

A cost of living index measures the relative cost of goods and services in different cities or countries. It is calculated by comparing the cost of a basket of goods and services, such as food, housing, transportation, and healthcare, in different locations. The index is usually expressed in bar chart along with different parametre , The cost of living index can be used to compare the affordability of different cities or countries and can be a useful tool for budgeting and financial planning.

an analysis is done to determine cost of living in various cites of usa to manage their expences along with aware cost of daily usage things 
such as food , appartment rent , transportation and healthcare in different location

# 1:- find the city where  restautrent is expensive  and also find the cities where restaurent cost in avg price 

```python
df_x1=df_usa.sort_values(by=['x1'],ascending =False).head(20)  # to sort the data in desending order 
```

```python
plt.bar(df_x1['city'],df_x1['x1'])    #plotting bar graph 
plt.xticks(rotation='vertical')
plt.title("Most expensive city")
plt.xlabel("city")
plt.ylabel("cost of rasturent in usd")
plt.show()
```

```python
df_avgx1=df_usa[df_usa['x1']<14].head(50)  # to find most affordable city where resturent cost in budget
df_avgx1
```

```python
plt.figure(figsize=(15,8))
plt.bar(df_avgx1['city'],df_avgx1['x1'])   # plotting bar graph of most affordable city
plt.xticks(rotation='vertical')
plt.title("Most affordable city")
plt.xlabel("city")
plt.ylabel("cost of rasturent in usd")
plt.show()
```

from above we find list of most affordable city of usa 

Conclusion :- From above, we have get the the plot of the most expensive city and most affordable city in respect of restaurent we can clearly see that lorain city is most expensive with cost 55 usd and most affordable city are cleveland , tempe , Hayward with cost less then avg cost which is 14 usd.

Conclusion :- Hence we have successfully visualize the required columns of the given dataset and we have got the idea of cost of restaurent present in different city.
