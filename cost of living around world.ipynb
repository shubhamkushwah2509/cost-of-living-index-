# find  top 10 country most expensive restaurent in world

```python
df1=df.groupby('country',as_index=False)['x1'].mean() # we use mean value to find avg cost in whole country
df1
```

```python
df1=df1.sort_values(by=['x1'],ascending=False).head(10) # we convert data into desending order 
df1
```

```python
plt.figure(figsize=(18,7))
plt.bar(df1['country'],df1['x1'],color='black')
plt.title("cost of meal in restaurent ")
plt.xlabel("country")
plt.ylabel("cost in usd")
```

from above we get most expensive restautent which is Turkmenistan followed by bermuda
